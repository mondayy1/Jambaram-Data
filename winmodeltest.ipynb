```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from pycaret.classification import *
import json
import os

df = pd.read_csv('/mnt/disk1/hojoong/matches/bone.csv', index_col=False)
root = '/mnt/disk1/hojoong/matches/complete'
files = os.listdir(root)
```

```python
for file in files:
    new_df = pd.read_csv(f'{root}/{file}', index_col=0)
    print(new_df)
    df = pd.concat([df, new_df])
    break
```

```python
df_train = df.iloc[:160]
df_test = df.iloc[160:]

exp_clf = setup(data = df, target = 'win', session_id=42, use_gpu=True)
```

```python
best_model = compare_models()
```

```python
x = df.iloc[:, :-2]
y = df['win']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=42)
y_train
```

```python
model = GradientBoostingClassifier(random_state=42, tree_method='gpu_hist')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print(model.predict_proba(x_test))
```
