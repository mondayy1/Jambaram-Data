```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from pycaret.classification import *
import json
import os

df = pd.read_csv('/mnt/disk1/hojoong/matches/bone.csv', index_col=False)
root = '/mnt/disk1/hojoong/matches/test'
files = os.listdir(root)
```

```python
def get_champid_from_participants(participants):
    
    #championids = []
    championids = {col:0 for col in df.columns}

    for participant in participants:
        #championids.append(participant['championId'])
        championids[str(participant['championId'])] = 1

    return championids
```

```python
for file in files:
    
    with open(f'{root}/{file}', 'r') as f:
        data = json.load(f)
    f.close()

    participants_blue = data['info']['participants'][:5]
    participants_red = data['info']['participants'][5:]

    championids_blue = get_champid_from_participants(participants_blue)
    championids_red = get_champid_from_participants(participants_red)

    if data['info']['teams'][0]['win'] == True: #Blue Win
        #championids_blue.append(1)
        #championids_red.append(0)
        championids_blue['win']=1
        championids_red['win']=0
    else: #Red Win
        #championids_blue.append(0)
        #championids_red.append(1)
        championids_blue['win']=0
        championids_red['win']=1
    
    df.loc[len(df)] = championids_blue
    df.loc[len(df)] = championids_red
df
```

```python
df_train = df.iloc[:160]
df_test = df.iloc[160:]

exp_clf = setup(data = df_train, target = 'win', session_id=42, use_gpu=True)
```

```python
best_model = compare_models()
```

```python
x = df.iloc[:, :-2]
y = df['win']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=42)
y_train
```

```python
model = GradientBoostingClassifier(random_state=42, tree_method='gpu_hist')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print(model.predict_proba(x_test))
```
